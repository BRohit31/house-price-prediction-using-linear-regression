## Splitting training and testing datasets

Specifying random_state to keep the test and training data separate even after randomizing to prevent overfitting

```python
train_set, test_set = train_test_split(house_data,test_size=0.2,random_state=42)
print(f"Rows in train set :{len(train_set)}\nRows in test set :{len(test_set)}")
```

Stratify to ensure similar distribution of data between test and training sets in case it is skewed in any way

```python
from  sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(house_data,house_data['CHAS']):
  strat_train_set = house_data.loc[train_index]
  strat_test_set = house_data.loc[test_index]

strat_test_set['CHAS'].value_counts()
```

## Missing attributes
Ways to deal with blank or inconsistent data

```python
house_data.describe()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy = 'median')
imputer.fit(house_data)
```

```python
imputer.statistics_
```

```python
housing_data_fit = imputer.transform(house_data)
housing_data_tr = pd.DataFrame(housing_data_fit, columns=house_data.columns)
```

```python
housing_data_tr.describe()
```

## Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

my_pipeline = Pipeline([
  ('imputer',SimpleImputer(strategy='median')),
     # ....add as many as reqd
  ('stdscaler',StandardScaler())
])
housing_num_tr = my_pipeline.fit_transform(housing_data_tr)
```

## Training and Testing the Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

model = LinearRegression()

model.fit(housing_num_tr,housing_labels)
```

```python
some_data=house_data.iloc[:5]
some_labels = housing_labels[:5]
prepared_data = my_pipeline.transform(some_data)
model.predict(prepared_data)
```

```python
list(some_labels)
```

```python
from sklearn.metrics import mean_squared_error
housing_prediction = model.predict(housing_num_tr)
lin_mse = mean_squared_error(housing_labels,housing_prediction)
lin_rmse = np.sqrt(lin_mse)
```

Using Cross Validation for evaluation

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model,housing_num_tr,housing_labels,scoring="neg_mean_squared_error")
rmse_scored = np.sqrt(-scores)
```

```python
def print_scores(scores):
  print("Scores :", scores)
  print("Mean : ", scores.mean())
  print("Standard deviation : ",scores.std())

print_scores(rmse_scored)
```

```python
X_test = strat_test_set.drop("MEDV",axis=1)
Y_test = strat_test_set['MEDV'].copy()
X_test_prepared = my_pipeline.transform(X_test)
final_predictions = model.predict(X_test_prepared)
final_mse = mean_squared_error(Y_test, final_predictions)
final_rmse = np.sqrt(final_mse)

print(final_rmse)
```
